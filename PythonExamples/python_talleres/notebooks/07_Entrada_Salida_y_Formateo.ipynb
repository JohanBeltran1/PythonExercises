# 7) Entrada/Salida y formateo

## ¿Qué es?
La **interacción por consola**: leer con `input()` y mostrar con `print()`.

## ¿Para qué se usa?
Para **recoger datos del usuario** y **presentar resultados** con buen formato.

## Claves de formateo
- f-strings: `f"{valor:.2f}"`
- `sep` y `end` en `print`
- Conversión explícita de tipos al leer (`int`, `float`)


```python
# Suma con formateo
a = 12.3456
b = 7.8
print(f"Suma: {a + b:.2f}")
print("uno", "dos", "tres", sep=" | ", end=".
")
```

### Ejercicios
- E1: Pide nombre y año de nacimiento; calcula edad.
- E2: Factura: subtotal, IVA (19%), total con 2 decimales.
- E3: “Ficha” multi-línea con `\n` y f-strings.

### Reto extra
Lee tres notas y muestra promedio, mínima, máxima y si aprueba (>=3.0).


```python
#ejercicio 1
from datetime import datetime

def normaliza_nombre(s: str) -> str:
    return s.strip().title()

def calcula_edad(anio_nacimiento: int) -> int:
    anio_actual = datetime.now().year
    return anio_actual - anio_nacimiento

nombre_input = input("Ingrese su nombre: ")
anio_input = int(input("Ingrese su año de nacimiento: "))

nombre = normaliza_nombre(nombre_input)

try:
    anio_nacimiento = int(anio_input)
    edad = calcula_edad(anio_nacimiento)
    print(f"Hola, {nombre}. Tienes {edad} años.")
except ValueError:
    print("El año de nacimiento debe ser un número válido.")

#ejercicio 2
def factura(subtotal: float) -> None:
    iva = subtotal * 0.21
    total = subtotal + iva
    print(f"Subtotal: {subtotal:.2f}")
    print(f"IVA (21%): {iva:.2f}")
    print(f"Total: {total:.2f}")

factura(10000)

#ejercicio 3
def mostrar_ficha(nombre: str, edad: str, ciudad: int) -> None:
    ficha = (
        f"Ficha de Usuario\n"
        f"----------------\n"
        f"Nombre : {nombre}\n"
        f"Edad  : {edad} años\n"
        f"Ciudad: {ciudad}\n"
    )
    print(ficha)

mostrar_ficha("Ethan Graves", "20", "New York")

#extra
def promedio(notas: list[float]) -> float:
    return sum(notas) / len(notas)

def lee_notas(n: int) -> list[float]:
    notas = []
    for i in range(1, n + 1):
        while True:
            try:
                nota = float(input(f"Ingrese la nota {i}: "))
                if 1.0 <= nota <= 7.0:
                    notas.append(nota)
                    break
                else:
                    print("La nota debe estar entre 1.0 y 7.0")
            except ValueError:
                print("Ingresa un número válido.")
    return notas

def mostrar_resultados(notas: list[float]) -> None:
    prom = promedio(notas)
    nota_min = min(notas)
    nota_max = max(notas)
    estado = "Aprobado" if prom >= 3.0 else "Reprobado"

    print("\n RESULTADOS")
    print(f"Notas  : {notas}")
    print(f"Promedio: {prom:.2f}")
    print(f"Mínima  : {nota_min}")
    print(f"Máxima  : {nota_max}")
    print(f"Estado  : {estado}")


notas = lee_notas(3)
mostrar_resultados(notas)
```
